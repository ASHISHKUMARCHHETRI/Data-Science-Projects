```python
data=pd.read_csv(r"C:\Mba notes\novels\500_Person_Gender_Height_Weight_Index.csv")
```

```python
plt.figure(figsize=(15,8))
df1=data.copy()
df1["Height"]=pd.cut(df1["Height"],[140,150,160,170,180,190,200],labels=["140-150","150-160","160-170","170-180","180-190","190-200"])
sns.boxplot(x="Height",y="Bmi",data=df1)
```

```python
sns.jointplot(x="Height",y="Bmi",data=data,ratio=3,color="y")
```

```python
df1=data.copy()
plt.figure(figsize=(12,8))
df1["Weight"]=pd.cut(df1["Weight"],[50,70,90,110,130,150,170],labels=["50-70","70-90","90-110","110-130","130-150","150-170"])
sns.pointplot(x="Weight",y="Bmi",data=df1,color="lime")
plt.xlabel("Weight",fontsize=14,color="yellow")
plt.ylabel("BMI",fontsize=14,color="green")
plt.title("Weight-Height Graph")
```

```python
df1=data.copy()
plt.figure(figsize=(12,8))
df1["Height"]=pd.cut(df1["Height"],[140,150,160,170,180,190,200],labels=["140-150","150-160","160-170","170-180","180-190","190-200"])
df1["Weight"]=pd.cut(df1["Weight"],[50,70,90,110,130,150,170],labels=["50-70","70-90","90-110","110-130","130-150","150-170"])
sns.pointplot(x="Height",y="Bmi",data=df1,color="blue",alpha=0.9)
sns.pointplot(x="Weight",y="Bmi",data=df1,color="black",alpha=0.5)
plt.text(6,0.8,"HEIGHT",color="blue",style="italic")
plt.text(6,0.8,"WEIGHT",color="black",style="italic")
plt.show()
```
