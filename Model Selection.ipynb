```python
from sklearn import datasets
from pandas import pandas as pd
from sklearn.model_selection import RandomizedSearchCV
```

```python
iris=datasets.load_iris()
```

```python
dir(iris)
```

```python
data=pd.DataFrame(iris.data,columns=iris.feature_names)
data["flower"]=iris.target
data
```

```python
data["flower"]=data["flower"].apply(lambda x:iris.target_names[x])
```

```python
data
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(iris.data,iris.target,test_size=0.2)
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
```

```python
clf=GridSearchCV(SVC(gamma="auto"),{
    "C":[1,10,20],
    "kernel":["rbf","linear"]
},cv=5,return_train_score=False)
clf.fit(iris.data,iris.target)
clf.cv_results_
```

```python
x_train,x_test,y_train,y_test=train_test_split(iris["data"],iris["target"],test_size=0.2)
rs=RandomizedSearchCV(LogisticRegression(max_iter=1000),{
    "C":[1,10,20],
    
},cv=5,return_train_score=False)
rs.fit(x_train,y_train)
rs.predict(x_test)
rs.score(x_test,y_test)
```

```python
model_params={
    "svm":{
        "model":SVC(gamma="auto"),
        "params":{
            "C":[1,10,20],
            "kernel":["rbf","linear"]
        }
    },
    "random_forest":{
        "model":RandomForestClassifier(),
        "params":{
            "n_estimators":[1,5,10]
        }
    },
    "logistic_regrssion":{
        "model":LogisticRegression(solver="liblinear",multi_class="auto"),
        "params":{
            "C":[1,5,10]
        }
    }
}

    
```

```python
scores=[]
for model_name,np in model_params.items():
    clf1=GridSearchCV(np["model"],np["params"],cv=5,return_train_score=False)
    clf1.fit(iris.data,iris.target)
    scores.append({
        "model":model_name,
        "best_sscore":clf1.best_score_,
        "best_params":clf1.best_params_
    })
```

```python
df=pd.DataFrame(scores,columns=["model","best_sscore","best_params"])
df
```
