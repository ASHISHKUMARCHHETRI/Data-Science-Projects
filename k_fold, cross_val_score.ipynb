```python
from sklearn.datasets import load_digits
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
digits=load_digits()
```

```python
dir(digits)
```

```python
data=pd.DataFrame(digits["data"])
```

```python
data["target"]=digits.target
```

```python
data
```

```python
x=data.drop("target",axis="columns")
```

```python
y=data["target"]
```

```python
x
```

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
```

```python
kf=KFold(n_splits=3)
```

```python
for a,b in kf.split([1,2,3,4,5,6,7,8,9]):
    print(a,b)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
def score(model,x_train,x_test,y_train,y_test):
    model.fit(x_train,y_train)
    print(model.score(x_test,y_test))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
```

```python
score(RandomForestClassifier(),x_train,x_test,y_train,y_test)
```

```python
score(LinearRegression(),x_train,x_test,y_train,y_test)
```

```python
score(SVC(),x_train,x_test,y_train,y_test)
```

```python
score(DecisionTreeClassifier(),x_train,x_test,y_train,y_test)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cross_val_score(RandomForestClassifier(),x,y)
```

```python
cross_val_score(DecisionTreeClassifier(),x,y)
```

```python
cross_val_score(LinearRegression(),x,y)
```

```python
cross_val_score(SVC(),x,y)
```
